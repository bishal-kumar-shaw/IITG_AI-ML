## Measure of Central Tendency

### Mean

The statistical mean refers to the mean or average that is used to derive the central tendency of the data. It is determined by adding all the data points in a population and then dividing the total by the number of points. The resulting number is known as the mean or the average.

Mathematically,
               $$Mean: \mu = \frac{1}{n}\sum_i^n{x_i} $$

```python
# Defining our own function to calculate Mean of a list
def mean(t):
    return(float(sum(t)) / len(t))
```

```python
mean([1,4,3,2,6,4,4,3,2,6,8,12,23])
```

```python
# Using Built-in library function for calculating Mean
import numpy as np
np.mean([1,4,3,2,6,4,4,3,2,6,8,12,23])
```

### Median

The median is a simple measure of central tendency. To find the median, we arrange the observations in order from smallest to largest value. If there is an odd number of observations, the median is the middle value. If there is an even number of observations, the median is the average of the two middle values.

Mathematically,
                $$ median(a) = a\lfloor{\frac{l+1}{2}}\rfloor + a\lceil{\frac{l+1}{2}}\rceil$$
                
where  __a__ is an ordered list of __l__ numbers, and ⌊ ⋅ ⌋ and ⌈ ⋅ ⌉ denote the floor and ceiling functions, respectively.

```python
# Defining our own function to calculate Median of a list
def median(lst):
    sortedLst = sorted(lst)
    lstLen = len(lst)
    index = (lstLen - 1) // 2 # floor division

    if (lstLen % 2 !=0):
        return sortedLst[index] 
    else:
        return (sortedLst[index] + sortedLst[index + 1])/2.0
```

```python
ls = [5,6,9,8]
print(sorted(ls))
print(median(ls))
```

```python
# Using Built-in library function for calculating Median
import statistics

ls = [5,6,9,8]

print(statistics.median(ls))
```

### Mode

The mode of a set of data values is the value that appears most often. If X is a discrete random variable, the mode is the value x (i.e, X = x) at which the probability mass function takes its maximum value. In other words, it is the value that is most likely to be sampled.

```python
# Defining our own function to calculate Mode of a list
import numpy as np
def mode(xs):
    values, counts = np.unique(xs, return_counts=True)
    max_count_index = np.argmax(counts) #return the index with max value counts
    return values[max_count_index]
```

```python
print(mode([1,7,2,5,3,3,8,3,2]))
```

```python
# Using Built-in library function for calculating Median
import statistics

ls = [1,7,2,5,3,3,8,3,2]

print(statistics.mode(ls))
```

### Variance

In probability theory and statistics, Variance is the expectation of the squared deviation of a random variable from its mean. Informally, it measures how far a set of (random) numbers are spread out from their average value.

Mathematically,
              $$Variance: \sigma^2 = \dfrac{1}{n}\sum_{i=1}^n (x_i - \mu)^2 $$

The $x_i – μ$ is called the “deviation from the mean”, making the Variance the squared deviation multiplied by 1 over the number of samples. This is why the square root of the variance, σ, is called the standard deviation.

```python
# Defining our own function to calculate variance of a list
def var(t, mu):
    dev2 = [(x - mu)**2 for x in t]
    var = np.mean(dev2)
    return(var)
```

```python
ls =[1,3,3,6,3,2,7,5,9,1]
print(var(ls,np.mean(ls)))
```

```python
# Using Built-in library function for calculating Varaince
import numpy as np
print(np.var([1,3,3,6,3,2,7,5,9,1]))
```

## Skewness (Third Order Moment)

Skewness is asymmetry in a statistical distribution, in which the curve appears distorted or skewed either to the left or to the right. Skewness can be quantified to define the extent to which a distribution differs from a normal distribution. Mathematically, it is denoted as
$$ Skew[X] = E\bigg[\bigg(\frac{X-\mu}{\sigma}\bigg)^{3}\bigg]$$

```python
from scipy.stats import skew 
import numpy as np  
import matplotlib.pyplot as plt  
  
x1 = np.linspace( -5, 5, 1000 ) 
y1 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2) 
  
plt.plot(x1, y1) 
  
print( '\nSkewness for data : ', skew(y1)) 
```

## Kurtosis (Fourth Order Moment)

Kurtosis is the fourth central moment divided by the square of the variance. If Fisher’s definition is used, then 3.0 is subtracted from the result to give 0.0 for a normal distribution. The expression for kurtosis is given by <br>

$$ Kurt[X] = E\bigg[\bigg(\frac{X-\mu}{\sigma}\bigg)^{4}\bigg]$$

```python
from scipy.stats import kurtosis 
import numpy as np  
import matplotlib.pyplot as plt  


mu_1 = 0
sig_1 = 1
x1 = np.linspace( -5, 5, 1000 ) 
y1 = 1./(np.sqrt(2.*np.pi)) * np.exp(-.5*((x1-mu_1)**2)/(sig_1**2))  
  
plt.plot(x1, y1, color='red') 

mu_2 = 0
sig_2 = 2
x2 = np.linspace( -5, 5, 1000 ) 
y2 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*((x2-mu_2)**2)/(sig_2**2))

plt.plot(x2,y2)

print( '\nKurtosis for the Red normal distribution :', kurtosis(y1,fisher=True)) 
  
print( '\nKurtosis for Blue normal distribution :', kurtosis(y2,fisher=True)) 
  
```

## Study of Different Statistical Distribution 

A random variable is a variable whose possible values are numerical outcomes of a random phenomenon. There are two types of random variables, __discrete__ and __continuous__.

A __discrete random variable__ is one which may take on only a countable number of distinct values and thus can be quantified. For example, you can define a random variable $X$ to be the number which comes up when you roll a fair dice. $X$ can take values : $[1,2,3,4,5,6]$ and therefore is a discrete random variable.

The probability distribution of a discrete random variable is a list of probabilities associated with each of its possible values. It is also sometimes called the probability function or the probability mass function. To have a mathematical sense, suppose a random variable $X$ may take $k$ different values, with the probability that $X = x_{i}$ defined to be $P(X = x_{i}) = p_{i}$. Then the probabilities $p_{i}$ must satisfy the following:

1: 0 < $p_{i}$ < 1 for each $i$

2: $p_{1} + p_{2} + ... + p_{k} = 1$.

Some examples of discrete probability distributions are __Bernoulli distribution__, __Binomial distribution__, __Poisson distribution__ etc.

---

A __continuous random variable__ is one which takes an infinite number of possible values. For example, you can define a random variable $X$ to be the height of students in a class. Since the continuous random variable is defined over an interval of values, it is represented by the area under a curve (or the integral).

The probability distribution of a continuous random variable, known as probability distribution functions, are the functions that take on continuous values. The probability of observing any single value is equal to $0$ since the number of values which may be assumed by the random variable is infinite. For example, a random variable $X$ may take all values over an interval of real numbers. Then the probability that $X$ is in the set of outcomes $A, P(A)$, is defined to be the area above $A$ and under a curve. The curve, which represents a function $p(x)$, must satisfy the following:

1: The curve has no negative values $(p(x) > 0$ for all $x$)

2: The total area under the curve is equal to $1$.

A curve meeting these requirements is often known as a density curve. Some examples of continuous probability distributions are __normal distribution__, __exponential distribution__, __beta distribution__, etc.

### Normal Distribution

The probability distribution function of a normal density curve with mean $\mu$ and standard deviation $\sigma$ is given by <br>
$$ f(x| \mu, \sigma^{2}) = \frac{1}{\sqrt{2\pi\sigma^{2}}}e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}}$$

```python
from scipy.stats import norm
# generate random numbers from N(0,1)
data_normal = norm.rvs(size=10000,loc=0,scale=1)
ax = sns.distplot(data_normal,
                  bins=100,
                  color='skyblue')
ax.set(xlabel='Normal Distribution', ylabel='Frequency')
```

### Gamma Distribution

The expression for gamma distribution is given by <br>

$$f(x;\alpha, \beta) = \frac{\beta^{\alpha}x^{\alpha-1}e^{-\beta x}}{\Gamma(a)}$$

Here $\alpha$ is the shape parameter and $\beta$ is the rate parameter

```python
from scipy.stats import gamma
data_gamma = gamma.rvs(a=5, size=10000)
ax = sns.distplot(data_gamma,
                  bins=100,
                  color='skyblue')
ax.set(xlabel='Gamma Distribution', ylabel='Frequency')
```

### Exponential Distribution

The expression for exponential distribution is given by <br>
$$ f(x;\lambda)=\begin{cases}
    \lambda e^{-\lambda x} \;, & \text{if $x \geq 0 $}.\\
    0, & \text{if $x < 0$} 
    \end{cases}$$ 
Here $\lambda$  is the rate parameter. 

```python
from scipy.stats import expon
data_expon = expon.rvs(scale=1,loc=0,size=1000)
ax = sns.distplot(data_expon,
                  bins=100,
                  color='skyblue')
ax.set(xlabel='Exponential Distribution', ylabel='Frequency')
```

### Poisson Distribution

Poisson distribution is described in terms of the rate $\mu$ at which the events happen. An event can occur 0, 1, 2, … times in an interval. The average number of events in an interval is designated $\lambda$ (lambda).
The probability of observing $k$ events in an interval is given by the equation <br>

P ($k$ events in the interval) = $ e^{-\lambda}\frac{{\lambda}^{k}}{k!}$

```python
from scipy.stats import poisson
data_poisson = poisson.rvs(mu=5, size=10000)
ax = sns.distplot(data_poisson,
                  bins=30,
                  kde=False,
                  color='skyblue')
ax.set(xlabel='Poisson Distribution', ylabel='Frequency')
```

### Binomial Distribution

A distribution where only two outcomes are possible, such as success or failure, gain or loss, win or lose and where the probability of success and failure is same for all the trials is called a Binomial Distribution. However, The outcomes need not be equally likely, and each trial is independent of each other. The parameters of a binomial distribution are n and p where n is the total number of trials, and p is the probability of success in each trial. Its probability distribution function is given by :

$$f(k,n,p) = Pr(k;n,p) = Pr(X=k) = \binom{n}{k} \cdot p^kq^{n-k} $$
where,
$$\binom{n}{k} = \frac{n!}{k!(n-k)!}$$

```python
from scipy.stats import binom
data_binom = binom.rvs(n=10,p=0.8,size=10000)
ax = sns.distplot(data_binom,
                  kde=False,
                  color='skyblue')
ax.set(xlabel='Binomial Distribution', ylabel='Frequency')
```

### Bernoulli Distribution

A Bernoulli distribution has only two possible outcomes, namely 1 (success) and 0 (failure), and a single trial, for example, a coin toss. So the random variable X which has a Bernoulli distribution can take value 1 with the probability of success, p, and the value 0 with the probability of failure, q or 1−p. The probabilities of success and failure need not be equally likely. The Bernoulli distribution is a special case of the binomial distribution where a single trial is conducted (n=1). Its probability mass function is given by:

$$ f(k;p) = p^k(1-p)^{(1-k)},\;\;\;\;\; for\; k \in \{0,1\}$$

```python
from scipy.stats import bernoulli
data_bern = bernoulli.rvs(size=10000,p=0.5)
ax= sns.distplot(data_bern,
                 kde=False,
                 color="skyblue")
ax.set(xlabel='Bernoulli Distribution', ylabel='Frequency')
```

## Mahalanobis Distance

### Mahalanobis Distance of a point from a cluster

Mahalanobis distance between a point and a cluster of points can be defined as : <br>
    <h1><center> $ D_{M}(\vec{x})  =  \sqrt{(\vec{x}-\vec{\mu})^{T}S^{-1}(\vec{x}-\vec{\mu})} $ </center></h1>
    Here, $\mu $ is the mean of the points in the cluster and $S$ is the covariance matrix of the set of points.

```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
def mahalanobisDist(data_vec,mean,cov,pointNum):
    plt.figure(figsize = (10,10))
    x,y = np.random.multivariate_normal(mean, cov, pointNum).T
    plt.scatter(mean[0],mean[1],marker="X",color="red")
    plt.scatter(data_vec[0],data_vec[1],marker="o",color="black")
    plt.scatter(x,y)
    
    temp = np.matmul((data_vec-mean),LA.inv(cov))
    dist = np.matmul(temp,(data_vec-mean).T)
    print(dist)
```

```python
mean = np.array([7,8])
cov  = np.array([[4,3],
                 [3,9]])
# cov = cov/10
data_vec = np.array([6,5]) 
mahalanobisDist(data_vec,mean,cov,100)
```

Mahalanobis distance between two clusters of points

The Mahalanobis distance between 2 clusters of points with the first cluster having mean and covariance matrix $\mu_{1}$ and $S_{1}$ and the second cluster having mean and covariance matrix $\mu_{2}$ and $S_{2}$ is given by <br>

 $$ D_{M}(C_{j},C_{i})  =  \dfrac{1}{2}(\vec{\mu_{j}}-\vec{\mu_{i}})^{T}(S_{j}^{-1}+S_{i}^{-1})(\vec{\mu_{j}}-\vec{\mu_{i}}) $$ 

```python
def mDistanceClust(clus1,clus2):
    u_i = np.mean(clus1,axis=1)
    S_i = np.cov(clus1)
    
    u_j = np.mean(clus2,axis=1)
    S_j = np.cov(clus2)
    
    temp1 = np.matmul((u_j-u_i).T,(LA.inv(S_j)+LA.inv(S_i)))
    temp2 = np.matmul(temp1,(u_j-u_i))
    d = 0.5*temp2
    
    return(d)
```

```python
mean1 = np.array([0,0])
cov1 = np.array([[3,0],[0,6]])
num1 = 100
clus1 = np.random.multivariate_normal(mean1, cov1, num1).T

mean2 = np.array([7,7])
cov2 = np.array([[2,0],[0,2]])
num2 = 100
clus2 = np.random.multivariate_normal(mean2, cov2, num2).T

plt.figure()
plt.scatter(clus1[0,:],clus1[1,:],color='red')
plt.scatter(clus2[0,:],clus2[1,:],color='blue')
plt.show()

## To show that the order of i and j does not matter
print("The distance between red cluster and blue cluster is {}".format(mDistanceClust(clus1,clus2)))
print("The distance between blue cluster and red cluster is {}".format(mDistanceClust(clus2,clus1)))
```
