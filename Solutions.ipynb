## Problem: Write a function to print all prime numbers in an interval [a,b]. Interval is to be obtained as a user input

```python
# Creating a user-defined function for checking prime numbers
def checkPrime(n):
    count = 0
    for i in range(2,n):
        if n%i == 0:
            count +=1
    if count!=0:
        return False
    else:
        return True   
```

```python
# Taking Intervals as Inputs from user 
a = int(input("Enter the lower bound: "))
b = int(input("Enter the upper bound: "))
```

```python
# Loop to interate form lower to upper bound
for i in range(a,b+1):
    if checkPrime(i) == True:
        print(i)
    else:
        continue
```

## Problem: User input three numbers a, b and c. Print their lowest common multiple(LCM)

```python
# To find LCM without finding GCD 
a = int(input("Enter the first number "))
b = int(input("Enter the second number "))
c = int(input("Enter the third number "))
Max = max(a,b,c)
for i in range(Max,a*b*c+1):
    if i%a==0 and i%b==0 and i%c==0:
        print("LCM: ",i)
        break
```

- Compute the following: $ \mathbf{C} = \frac{1}{n}\sum_{i=1}^{n}\big(x_{i}-\mu\big)\big(x_{i}-\mu\big)^{T}$, where $\mu =\frac{1}{n}\sum_{i=1}^{n}x_{i}, i=1,2,...,n$, $x_{i} = \big[x_{i1},x_{i2},...,x_{i10}\big]^{T}$ is a column vector.

```python
# Reading myCSV.csv into numpy array and computing the co-variance matrix
Data = np.genfromtxt("myCSV.csv",delimiter=",")
Data
```

```python
l = len(Data)
c = np.zeros((10,10))
mu = np.mean(Data,axis=0)
for i in range(l): # we are taking data point as a Row  
    c += np.matmul(np.matrix((Data[i,:]-mu)).T,np.matrix((Data[i,:]-mu)))
C = c/l
print(C)
```

## Problem : Define a class for a complex number $a + jb$ . Define memeber functions to do basic operations conjugate, absolute value, addition, subtraction,multiplication, division and angle. Define two complex numbers c1 , c2 and print the results of the following operations $c1+c2 ,c1-c2 ,c1*c2 , c1/c2 , |c1| , |c2|, \angle c1, \angle c2 $.

```python
# We have to built our own class for this problem
import math
class complex:
    real = 0
    img = 0
    def __init__(self,Real,Img): #constructor 
        self.real = Real
        self.img = Img
    def compAdd(c1,c2):
        c = complex(0,0)
        c.real = c1.real + c2.real
        c.img = c1.img + c2.img
        return c
    def compSub(c1,c2):
        c = complex(0,0)
        c.real = c1.real - c2.real
        c.img = c1.img - c2.img
        return c
    def compMul(c1,c2):
        c = complex(0,0)
        c.real = c1.real*c2.real - c1.img*c2.img
        c.img = c1.img*c2.real + c1.real*c2.img
        return c
    def compDivide(c1,c2):
        c = complex(0,0)
        c.real = (c1.real*c2.real + c1.img*c2.img)/(c2.real**2+c2.img**2)
        c.img = c1.img*c2.real - c1.real*c2.img/(c2.real**2+c2.img**2)
        return c
    def compAbs(c):
        return math.sqrt(c.real**2+c.img**2)
    def compArg(c):
        return math.atan(c.img/c.real)
    def compPrint(c):
        print(str(c.real)+"+j("+str(c.img)+")")
```

```python
c1 = complex(3,5)
c2 = complex(4,6)
complex.compPrint(complex.compAdd(c1,c2))
complex.compPrint(complex.compSub(c1,c2))
complex.compPrint(complex.compMul(c1,c2))
complex.compPrint(complex.compDivide(c1,c2))
print(complex.compAbs(c1))
print(complex.compAbs(c2))
print(complex.compArg(c1))
print(complex.compArg(c2))
```

## Problem: Scatter Plot 
- Generate a set of n = 100 points, $X = \{x_{i}\}, i=1,2,...,n, x_{i}\in\mathcal{R^{2}}$ within an ellipse $\frac{\big(x-\mu_x\big)^{2}}{a^{2}} + \frac{\big(y-\mu_y\big)^{2}}{b^{2}} = 1$ centered at $\big[\mu_x,\mu_y\big] = [5,-5]$ and has a major axis $2a = 10$ and minor axis $2b = 5$
- Scatter plot all the points.

```python
import numpy as np
from matplotlib import pyplot as plt
from math import pi

u=5     #x-position of the center
v=-5    #y-position of the center
a=5     #radius on the x-axis
b=2.5    #radius on the y-axis

t = np.linspace(0, 2*pi, 100)
plt.figure()
plt.plot( u+a*np.cos(t) , v+b*np.sin(t) )
plt.grid(color='lightgray',linestyle='--')
points = 0
xarr = []
yarr = []
while points<100:
    theta = np.random.uniform(0,2*pi)
    x = u+np.random.uniform(0,a)*np.cos(theta)
    y = v+np.random.uniform(0,b)*np.sin(theta)
    if ((x-u)/a)**2 + ((y-v)/b)**2 < 1:
        points +=1
        xarr.append(x)
        yarr.append(y)
plt.scatter(xarr,yarr,color='red')        
plt.show()
```
